```python
# Updating categories so that "Overall" applies only to accuracy
classes = ["Undisturbed Forest", "Disturbed Forest", "Overall Accuracy"]
precision_metrics = [0.96, 0.88, None]  # Precision only for specific classes
recall_metrics = [0.94, 0.92, None]  # Recall only for specific classes
f1_score_metrics = [0.95, 0.90, None]  # F1-score only for specific classes
accuracy_metrics = [None, None, 0.93]  # Accuracy only applies to the overall category

# Set up the x-axis positions and bar width
x = np.arange(len(classes))
width = 0.2

# Create the plot with the new arrangement
plt.figure(figsize=(10, 6))
bars1 = plt.bar(x - width, [m if m is not None else 0 for m in precision_metrics], width, label='Precision', color='blue', alpha=0.7)
bars2 = plt.bar(x, [m if m is not None else 0 for m in recall_metrics], width, label='Recall', color='green', alpha=0.7)
bars3 = plt.bar(x + width, [m if m is not None else 0 for m in f1_score_metrics], width, label='F1-Score', color='orange', alpha=0.7)
bars4 = plt.bar(x, [m if m is not None else 0 for m in accuracy_metrics], width, label='Accuracy', color='red', alpha=0.7)  # Accuracy on "Overall" only

# Add percentages on the bars
for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only label bars with a height
            plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f"{height:.2f}", ha='center', va='bottom')

# Set labels and titles
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.title('Classification Metrics for Undisturbed and Disturbed Forests')
plt.xticks(ticks=x, labels=classes)
plt.ylim(0, 1.1)
plt.legend()

# Save the updated graph
plt.tight_layout()
plt.savefig('disturbance_metrics', dpi=300)

# Display the graph and provide the download link
plt.show()

```

```python
import os
import glob
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Define the path to your folder containing .tif images
data_path = "F:/Disturbance"

# Define pixel area (adjust based on your dataset)
pixel_size = 10  # e.g., for Sentinel-2, 10m x 10m
pixel_area = pixel_size ** 2  # in square meters

# Get list of .tif files
tif_files = sorted(glob.glob(os.path.join(data_path, "predicted_core*.tif")))

# Extract years from filenames
def extract_year(filename):
    return int(os.path.basename(filename).split("_")[-1].split(".")[0])

# Store results
yearly_destruction = {}

for tif in tif_files:
    year = extract_year(tif)
    with rasterio.open(tif) as src:
        data = src.read(1)  # Read first band
        destruction_pixels = np.sum(data == 1)  # Count destroyed pixels
        destruction_area = destruction_pixels * pixel_area / 1e6  # Convert to square km
        yearly_destruction[year] = destruction_area

# Sort results by year
years = sorted(yearly_destruction.keys())
areas = [yearly_destruction[year] for year in years]

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(years, areas, marker="o", linestyle="-", color="r", label="Disturbed Forest Area")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Disturbed Area (km²)", fontsize=12)
plt.title("Forest Disturbance Over Time", fontsize=14)
# plt.grid(False, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig('disturbed forest_over_time', dpi=300)
plt.show()
```
